## Analysis for user 11
In Model Overview we do a recommendation for user 11. Here we can see what data is given for recommendations

```python
import pandas as pd
import json

with open('../generated_data/user_watched.json', 'r') as file:
    content = file.read()
json_objects = []


for line in content.splitlines():
    if line.strip():  # Ignore empty lines
        try:
            json_objects.append(json.loads(line.strip()))
        except json.JSONDecodeError as e:
            print(f"Skipping invalid JSON: {e}")

selected_user_data = next(
    (user for user in json_objects if user['user_id'] == 11), None)
if selected_user_data:
    anime_ratings = selected_user_data['anime_ratings']
    print(f"Structure of anime_ratings: {anime_ratings[:3]}")
    rows = []
    for anime in anime_ratings:
        if isinstance(anime, list):
            anime = {'anime_id': anime[0], 'name': anime[1],
                     'rating': anime[2], 'genre': anime[3]}

        genres = anime['genre'].split(',')
        rows.append({
            'anime_id': anime['anime_id'],
            'name': anime['name'],
            'rating': anime['rating'],
            'genres': ', '.join([genre.strip() for genre in genres])
        })

    df = pd.DataFrame(rows)
    sorted_df = df.sort_values(
        by=['rating', 'anime_id'], ascending=[False, True])
    pd.set_option('display.max_rows', None)
    pd.set_option('display.max_columns', None)
    pd.set_option('display.width', 1000)
    pd.set_option('display.max_colwidth', 1000)
    print("Sorted anime ratings for user:")
    print(sorted_df[['anime_id', 'name', 'rating', 'genres']])
    genre_rows = []
    for anime in anime_ratings:
        if isinstance(anime, list):
            anime = {'anime_id': anime[0], 'name': anime[1],
                     'rating': anime[2], 'genre': anime[3]}

        genres = anime['genre'].split(',')
        for genre in genres:
            genre_rows.append({
                'anime_id': anime['anime_id'],
                'rating': anime['rating'],
                'genre': genre.strip()
            })

    genre_df = pd.DataFrame(genre_rows)
    genre_averages = genre_df.groupby('genre')['rating'].mean().reset_index()
    genre_averages = genre_averages.sort_values(by='rating', ascending=False)
    print("\nGenre averages for user:")
    print(genre_averages[['genre', 'rating']])
else:
    print("User not found.")
```
